```python
#question 2
import pandas as pd

df = pd.read_csv("Inpatient_Rehabilitation_Facility-General_Information_Sep2024 - Inpatient_Rehabilitation_Facility-General_Information_Sep2024.csv - Inpatient_Rehabilitation_Facility-General_Information_Sep2024 - Inpatient_Rehab.csv")
print(df.head())
```

```python
#question 3
df['Certification Date'] = pd.to_datetime(df['Certification Date'], format='mixed')
non_profit = df[(df['Ownership Type'] == 'Non-profit') & (df['Certification Date'] > '2011-10-01')]
print(non_profit)
```

```python
#question 4
df['Certification Date'] = pd.to_datetime(df['Certification Date'], format='mixed')
birmingham_providers = df[(df['Ownership Type'] == 'For profit') & (df['City/Town'] == 'BIRMINGHAM')]
print(birmingham_providers)
```

```python
#question 5
zip_code = df[(df['ZIP Code'] >= 85000) & (df['ZIP Code'] <= 90000)]
print(zip_code)
```

```python
#question 6
count = df['Ownership Type'].value_counts()
profit_non_profit = count[['For profit', 'Non-profit']]
print(profit_non_profit)
```
